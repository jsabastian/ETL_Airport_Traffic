```python
# Import dependencies and Libraries
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import the raw US Airports data file as DataFrame
csv_file = "airports.csv"
airports_df = pd.read_csv(csv_file)
airports_df.head()
```

```python
# Extract only required columns and create a new dataframe
airports_df = airports_df[["IATA_CODE","AIRPORT","CITY","STATE","LATITUDE","LONGITUDE"]]
airports_df.head()
```

```python
def isnull_sum(df):
    return df.isnull().sum() > 0
```

```python
isnull_sum(airports_df)
```

```python
airports_df_1 = airports_df.dropna()
```

```python
airports_df_1 = airports_df_1.rename(columns = {"AIRPORT":"airport","AIRLINE":"airline",\
                                                "IATA_CODE":"iata_code","CITY":"city","STATE":"state_",\
                                                "LATITUDE":"latitude","LONGITUDE":"longitude"})
```

```python
airports_df_1.head()
```

```python
#import the 2nd airport data
csv_file_1 = "busiestAirports.csv"
usa_df = pd.read_csv(csv_file_1)
usa_df.head()
```

```python
isnull_sum(usa_df)
```

```python
usa_df = usa_df.rename(columns = {"total\r\npassengers":"total_passengers"})
usa_df.head()
```

```python
# Prepping to extract the data related to USA only
usa_df.count()
```

```python
# Extract only required columns and create a new dataframe
usa_selected_df = usa_df[["id","rank_","year_","iata_code","country","total_passengers"]]
usa_selected_df.head()
```

```python
is_usa =  usa_selected_df['country']== "United States"
print(is_usa.head())
```

```python
united_states_df = usa_selected_df[is_usa]
print(united_states_df)
```

```python
# Connect to Local Postgres Database server
#rds_connection_string = "<insert user name>:<insert password>@localhost:5432/customer_db"
rds_connection_string = "postgres:Catch-22@localhost:5432/flights"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Check for 4 Tables as they should exist and be empty
engine.table_names()
```

```python
#load csv into db
airports_df_1.to_sql(name = "airports", con=engine, if_exists="append", index = False)
```

```python
#check if data has been added
pd.read_sql_query('select * from airports', con=engine).head()
```

```python
#load csv into db
usa_selected_df.to_sql(name = "busiest", con=engine, if_exists="append", index = False)
```

```python
#check if data has been added
pd.read_sql_query('select * from busiest', con=engine).head()
```
